```python
import matplotlib.pyplot as plt
age_groups = ['0-20 Years', '21-64 Years', '65+ Years']
population_mn = [512, 807, 98]
percentages = [36.1, 57.0, 6.9]
colors = ['#ffe600', '#0099ff', '#ff4da6']
plt.figure(figsize=(8, 6))
bars = plt.bar(age_groups, population_mn, color=colors)
for bar, pct in zip(bars, percentages):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 10, f'{pct}%', ha='center', fontsize=12, fontweight='bold')
plt.title("India's Population Distribution by Age in 2022", fontsize=14, fontweight='bold')
plt.ylabel('Population (in Millions)', fontsize=12)
plt.xlabel('Age Groups', fontsize=12)
plt.ylim(0, max(population_mn) + 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
